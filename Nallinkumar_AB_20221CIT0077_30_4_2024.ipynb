```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
df=df.set_index('Date')
df['Open'].plot(linewidth=0.5);
```

```python
cols_plot=['High','Low','Open']
axes=df[cols_plot].plot(marker='.',alpha=0.5,linestyle='None',figsize=(11,9),subplots=True)
for ax in axes:
  ax.set_ylabel('Price ($)')
```

```python
fig, ax=plt.subplots()
ax.plot(df.loc['2021-01-01 23:59:59':,'Open'],marker='.',linestyle='-',linewidth=0.5,label='Daily')
ax.set_ylabel('Price ($)')
ax.legend();
```

```python
fig, ax=plt.subplots()
ax.plot(df.loc['2021-01-01':,'Open'],marker='o',markersize=8,linestyle='-',label='Weekly Mean Resample')
ax.set_ylabel('Price ($)')
ax.legend();
```
