```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x= np.linspace(0,108,100)
ax.plot(x,np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
x= np.linspace(0,54,100)
ax.plot(x,np.sin(x))
print(x)
```

```python
fig = plt.figure()
ax = plt.axes()

x= np.linspace(0,54,50)
ax.plot(x,np.sin(x))

ax.plot(x,np.cos(x))
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x= np.linspace(0,105,10)
ax.plot(x,np.sin(x),label='sin',color ='green')
ax.plot(x,np.cos(x),label='cos',color ='red')
ax.plot(x,np.tan(x),label='tan')
ax.set_title('Multiple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
```

```python
fig = plt.figure()
ax = plt.axes()

x= np.linspace(0,105,10)
ax.plot(x,np.sin(x),label='sin', linestyle ='dashed')
ax.plot(x,np.cos(x),label='cos', linestyle ='dotted')
ax.plot(x,np.tan(x),label='tan',linestyle ='dashdot')
ax.set_title('Multiple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```
